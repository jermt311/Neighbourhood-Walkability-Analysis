# Neighbourhood Walkability Analysis

## Project Introduction
In this project, several datasets were gathered to perform data anlaysis of walkability of different neighbourhoods in Sydney.

## Data Set Description and Preperation

Tha aim of this project is to calculate the `walkability score` for different neighbourhoods in Sydney similar to the 'WalkScore' service (https://www.walkscore.com/AU-NSW/Sydney). Walkability is commanly expressed as a measure of density, diversity, design, destination accessibility and distance to transport. 

To calculate this score, we will be integrating different data sources and based on the computed walkability scores, perform a correlation analysis with the median income and the average monthly rent per naighbourhood according to the provided Census Data from the Australian Bureau of Statistics (ABS).

### Connect to the server

Below, i removed my username and password for my privacy reasons.

```python
def pgconnect():
    # please replace <your_unikey> and <your_SID> with your own details
    YOUR_KEY = #'<your_key>'
    YOUR_PW     = #'<your_ID>'
    try: 
        conn = psycopg2.connect(host=,
                                database= YOUR_KEY,
                                user=YOUR_EY, 
                                password=YOUR_PW)
        
        print('connected')
    except Exception as e:
        print("unable to connect to the database")
        print(e)
        return None
    return conn

#check connection
conn = pgconnect()
```

```python
def pgquery( conn, sqlcmd, args=None, msg=False, returntype='tuple'):
    """ utility function to execute some SQL query statement
        it can take optional arguments (as a dictionary) to fill in for placeholders in the SQL
        will return the complete query result as return value - or in case of error: None
        error and transaction handling built-in (by using the 'with' clauses)"""
    retval = None
    with conn:
        cursortype = None if returntype != 'dict' else psycopg2.extras.RealDictCursor
        with conn.cursor(cursor_factory=cursortype) as cur:
            try:
                if args is None:
                    cur.execute(sqlcmd)
                else:
                    cur.execute(sqlcmd, args)
                if (cur.description != None ):
                    retval = cur.fetchall() # we use fetchall() as we expect only _small_ query results
                if msg != False:
                    print("success: " + msg)
            except psycopg2.DatabaseError as e:
                if e.pgcode != None and msg:
                    print("db read error: "+msg)
                    print(e)
            except Exception as e:
                print(e)
    return retval
```

### Neighbourhoods

### 1. Load Shapefile data

```python
import shapefile
df = pd.read_csv('Neighbourhoods.csv')
sf = shapefile.Reader("ESRI/AUS.shp", encoding="iso-8859-1")
print(len(sf.shapes()))
```

```python
sf.fields[1:]
```

### 2. Clean Neighbourhoods CSV file

```python
# check number of na values
print (df.isnull().sum())
# set na to 0
df['population'].fillna(0, inplace=True)
df['number_of_businesses'].fillna(0, inplace=True)
print (df.isnull().sum())
# export the cleaned csv file
df.to_csv('Neighbourhoods-cleaned.csv', index=False)

# read cleaned csv file
data = list(csv.DictReader(open('Neighbourhoods-cleaned.csv')))
```

### 3. Create Schema

```python
# Reset the table
pgquery (conn, "DROP TABLE IF EXISTS Neighbourhoods_Geo", None, "Reset Table Neighbourhoods_Geo")

neighbour_schema = '''CREATE TABLE IF NOT EXISTS Neighbourhoods_Geo (
                     area_id NUMERIC PRIMARY KEY,
                     area_name VARCHAR(50),
                     land_area NUMERIC,
                     population NUMERIC,
                     number_of_dwellings NUMERIC,
                     number_of_businesses NUMERIC,
                     geom GEOMETRY(Polygon,4283))''' 

# pgquery(conn, "DROP TABLE Assignment.Neighbourhoods_Geo", msg="cleared old table")
pgquery(conn, neighbour_schema, msg="created Neighbourhoods_Geo table")
```

### 4. Insert into table

**4.1 Extract geometry for 'Greater Sydney'**

```python
import re

shapes = sf.shapes()
records= sf.records()

# row = {}
geoms = []
for i in range(0, len(shapes)):
    record = sf.record(i)
    shape  = sf.shape(i)

    
    # prepare the polygon data
    # this is a bit complex with our dataset as it has complex polygons, some with multiple parts...
    if record[8] == 'Greater Sydney':
        geom_info = "POLYGON(("
        i=0
        for x, y in shape.points:
            geom_info += "%s %s," %(x, y)
           # check for start of a new polygon part
            i += 1
            if i in shape.parts:
                geom_info = re.sub(",$", "),(", geom_info)
        # properly end the polygon string
        geom_info = re.sub(",$", "))", geom_info)
        geoms.append(geom_info)
print(len(geoms))
print(len(data))

    
```

**4.2 Insert into table**

```python
insert_stmt = """INSERT INTO Neighbourhoods_Geo VALUES ( %(area_id)s, %(area_name)s, 
                    %(land_area)s, %(population)s, %(number_of_dwellings)s, 
                    %(number_of_businesses)s, ST_GEOMFROMTEXT(%(geom)s, 4283) )"""

row = {}
for i in range(len(data)):
    temp = data[i]
    row['area_id'] = temp['area_id']
    row['area_name'] = temp['area_name']
    row['land_area'] = temp['land_area']
    row['population'] = temp['population']
    row['number_of_dwellings'] = temp['number_of_dwellings']
    row['number_of_businesses'] = temp['number_of_businesses']
    row['geom'] = geoms[i]
    
#     finally: insert new row into the table
    pgquery(conn, insert_stmt, args=row, msg="inserted "+str(temp['area_name']))
    
```

## Compute Cyclability

```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

**1. Read file**

```python
df_ = pd.read_csv('compute_corr.csv')
```

**2. Clean data**

```python
df_['num_bikes'].fillna(0, inplace=True)
df_['num_scooters'].fillna(0, inplace=True)
df_['bike_density'].fillna(0, inplace=True)
df_
```

**3. Calculate density and diversity**

```python
df_['z_servicebalance'] = (df_.servicebalance - df_.servicebalance.mean())/df_.servicebalance.std(ddof=0)
df_['z_population_density'] = (df_.population_density - df_.population_density.mean())/df_.population_density.std(ddof=0)
df_['z_dwelling_density'] = (df_.dwelling_density - df_.dwelling_density.mean())/df_.dwelling_density.std(ddof=0)
df_['z_bike_density'] = (df_.bike_density - df_.bike_density.mean())/df_.bike_density.std(ddof=0)
df_['z_pods_density'] = (df_.pods_density - df_.pods_density.mean())/df_.pods_density.std(ddof=0)
df_['z_vehicle_density'] = (df_.vehicle_density - df_.vehicle_density.mean()) / df_.vehicle_density.std(ddof=0)
```

**4. Calculate cyclability**

```python
df_['cyclability']=-1*df_['z_population_density'] +df_['z_dwelling_density']+ df_['z_servicebalance']+df_['z_bike_density']+df_['z_pods_density']-df_['z_vehicle_density']
df_
```

```python
plt.hist(df_['cyclability'])
plt.show()
```

**5. Compute log of cyclability**

```python
for value in df_['cyclability']:
    if value >0:
        df_['log_cyclability']=np.log(df_['cyclability'])
    elif value == 0:
        df_['log_cyclability'] =0
    else:
        df_['log_cyclability']=-np.log(abs(df_['cyclability']))

df_
```

```python
plt.hist(df_['log_cyclability'])
plt.show()
```

**6. Normalise the cyclability score to 0-100**

```python
df_['scaled_log_cyclability']= 100*(df_['log_cyclability'] -df_['log_cyclability'].min())/(df_['log_cyclability'].max() -df_['log_cyclability'].min())
df_
```
